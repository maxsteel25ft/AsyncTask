```python
import numpy as np
import matplotlib.pyplot as plt

def generate_triangular_wave(frequency, amplitude, start, stop):
  """
  Generates samples of a triangular wave with user-specified parameters.

  Args:
    frequency: Frequency of the wave in Hz.
    amplitude: Amplitude of the wave.
    start: Starting time in seconds.
    stop: Ending time (duration) in seconds.

  Returns:
    time_samples: Array of time samples in seconds.
    triangular_wave: Array of corresponding triangular wave values.
  """

  # Calculate period
  period = 1 / frequency
  # Calculate total duration
  duration = stop - start

  # Ensure duration covers at least 3 cycles
  if duration < 3 * period:
    print("Warning: Duration is less than 3 cycles. Adjusting to 3 cycles.")
    duration = 3 * period

  # Number of samples
  num_samples = int(duration * sampling_rate)  # Adjust sampling_rate if needed

  # Generate time samples
  time_samples = np.linspace(start, stop, num_samples)

  # Generate triangular wave using piecewise linear function
  triangular_wave = np.piecewise(
      time_samples,
      [(time_samples < start + period/2), (time_samples >= start + period/2)],
      [lambda t: (2 * amplitude / period) * (t - start), lambda t: -2 * amplitude / period * (t - stop)]
  )

  return time_samples, triangular_wave

# Get user input
frequency = float(input("Enter frequency (Hz): "))
amplitude = float(input("Enter amplitude: "))
start_time = float(input("Enter starting time (seconds): "))
stop_time = float(input("Enter stopping time (duration, seconds): "))

# Generate triangular wave
time_samples, triangular_wave = generate_triangular_wave(frequency, amplitude, start_time, stop_time)

# Plot the triangular wave
plt.plot(time_samples, triangular_wave)
plt.grid()
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Triangular Wave (3 Cycles)')

# Adjust plot limits to show all 3 cycles (if necessary)
if stop_time - start_time < 3 * (1 / frequency):
  plt.xlim(start_time, start_time + 3 * (1 / frequency))  # Adjust x-axis if duration is less than 3 cycles

plt.ylim(-amplitude - 0.1, amplitude + 0.1)  # Adjust y-axis limits if needed

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_triangular_wave(frequency, amplitude, y1, y2, start, stop):
  """
  Generates samples of a triangular wave with user-specified parameters.

  Args:
    frequency: Frequency of the wave in Hz.
    amplitude: Amplitude of the variation from the base value (y1).
    y1: Starting value (base) of the triangular wave on the y-axis.
    y2: Peak value (top) of the triangular wave on the y-axis.
    start: Starting time in seconds.
    stop: Ending time (duration) in seconds.

  Returns:
    time_samples: Array of time samples in seconds.
    triangular_wave: Array of corresponding triangular wave values.
  """

  # Calculate period
  period = 1 / frequency
  # Calculate total duration
  duration = stop - start

  # Ensure duration covers at least 3 cycles
  if duration < 3 * period:
    print("Warning: Duration is less than 3 cycles. Adjusting to 3 cycles.")
    duration = 3 * period

  # Number of samples
  num_samples = int(duration * sampling_rate)  # Adjust sampling_rate if needed

  # Generate time samples
  time_samples = np.linspace(start, stop, num_samples)

  # Calculate delta values for slope and amplitude
  delta_y = y2 - y1
  delta_x = period / 2

  # Calculate slope (m)
  slope = delta_y / delta_x

  # Define piecewise linear function for the triangular wave
  def triangular_wave_func(t):
    if t < start + period/2:
      return slope * (t - start) + y1
    else:
      return -slope * (t - stop) + y2

  # Generate triangular wave values
  triangular_wave = triangular_wave_func(time_samples)

  return time_samples, triangular_wave

# Get user input
frequency = float(input("Enter frequency (Hz): "))
amplitude = float(input("Enter amplitude (variation from base value): "))
y1 = float(input("Enter base value (y1) of the wave: "))
y2 = float(input("Enter peak value (y2) of the wave: "))
start_time = float(input("Enter starting time (seconds): "))
stop_time = float(input("Enter stopping time (duration, seconds): "))

# Generate triangular waves for 3 cycles
time_samples_list = []
triangular_wave_list = []
for _ in range(3):  # Generate data for 3 cycles
  time_samples, triangular_wave = generate_triangular_wave(frequency, amplitude, y1, y2, start_time, stop_time)
  time_samples_list.append(time_samples)
  triangular_wave_list.append(triangular_wave)

# Plot the triangular waves (overlayed)
plt.plot(time_samples_list[0], triangular_wave_list[0], label='Cycle 1')  # Plot first cycle for reference
plt.plot(time_samples_list[1], triangular_wave_list[1], label='Cycle 2')
plt.plot(time_samples_list[2], triangular_wave_list[2], label='Cycle 3')
plt.grid()
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Triangular Waves (3 Cycles)')

# Adjust plot limits to show all 3 cycles (if necessary)
if stop_time - start_time < 3 * (1 / frequency):
  plt.xlim(start_time, start_time + 3 * (1 / frequency))  # Adjust x-axis if duration is less than 3 cycles

plt.ylim(min(y1 - amplitude, min(triangular_wave_list)) - 0.1, max(y2 + amplitude, max(triangular_wave_list)) + 0.1)  # Adjust y-axis to fit all waves

plt.legend()  # Add legend for each cycle
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def generate_staircase_waveform(Ei, Ef, deltaE, duration):
  """
  Generates a staircase voltage waveform with user-specified parameters.

  Args:
    Ei: Initial voltage (V).
    Ef: Final voltage (V).
    deltaE: Voltage increment per step (V/s).
    duration: Total duration of the waveform (s).

  Returns:
    time_samples: Array of time samples in seconds.
    voltage_steps: Array of corresponding voltage values at each step.
  """

  # Calculate the number of steps based on initial, final voltage, and increment
  num_steps = int((Ef - Ei) / deltaE) + 1  # Add 1 for final step

  # Ensure duration covers all steps
  if duration < num_steps * deltaE:
    print("Warning: Duration is less than the total staircase time. Adjusting duration.")
    duration = num_steps * deltaE

  # Generate time samples for each step edge (start and end of each step)
  time_edges = np.linspace(0, duration, num_steps + 1)

  # Generate voltage steps based on increment
  voltage_steps = Ei + np.arange(num_steps) * deltaE

  # Create time samples for each horizontal level (repeated for each voltage step)
  time_samples = np.repeat(time_edges[:-1], num_steps)

  # Combine voltage steps with repeated values for each level
  voltage_waveform = np.tile(voltage_steps, num_steps)

  return time_samples, voltage_waveform

# Get user input
Ei = float(input("Enter initial voltage (V): "))
Ef = float(input("Enter final voltage (V): "))
deltaE = float(input("Enter voltage increment per step (V/s): "))
duration = float(input("Enter total duration of the waveform (s): "))

# Generate staircase waveform
time_samples, voltage_waveform = generate_staircase_waveform(Ei, Ef, deltaE, duration)

# Plot the staircase waveform
plt.plot(time_samples, voltage_waveform, marker='o', linestyle='-', label='Staircase Waveform')
plt.grid(True)
plt.xlabel('Time (seconds)')
plt.ylabel('Voltage (V)')
plt.title('Staircase Voltage Waveform (Voltammetry)')

# Adjust plot limits to show entire duration and voltage range
plt.xlim(0, duration)
plt.ylim(Ei - deltaE, Ef + deltaE)  # Add some buffer to y-axis limits

plt.legend()
plt.show()
```
