```
import numpy as np

def generate_sine_wave(amplitude=4095, frequency=1.0, sample_rate=48):
  """
  This function generates an array containing a sinusoidal waveform for one full period.

  Args:
      amplitude (int, optional): The peak-to-peak amplitude of the waveform (default 4095).
      frequency (float, optional): The frequency of the sine wave in Hz (default 1.0).
      sample_rate (int, optional): The sampling rate in Hz (default 48).

  Returns:
      numpy.ndarray: An array containing the sine wave values.
  """
  # Define time points for one period
  t = np.linspace(0, 1, sample_rate)

  # Generate sine wave with amplitude scaling and offset
  waveform = amplitude * np.sin(2 * np.pi * frequency * t)

  # Scale waveform to range between 0 and amplitude (avoid -amplitude)
  waveform = (waveform + amplitude) / 2

  return waveform.astype(int)  # Convert to integer for DAC output

# Example usage with test plot
amplitude = 4095
frequency = 1.0
sample_rate = 48

sine_wave = generate_sine_wave(amplitude, frequency, sample_rate)
print(sine_wave.astype)
# Plot the waveform
import matplotlib.pyplot as plt
plt.plot(sine_wave)
plt.xlabel("Time Samples")
plt.ylabel("Amplitude (Integer)")
plt.title("Sine Waveform (One Period)")
plt.grid(True)
plt.show()
```
